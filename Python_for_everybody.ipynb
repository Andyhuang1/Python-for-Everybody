# 4. 字典Dictionaries

```python
# 字典 dict[key] = value  {key1:value1, key2:value2}
purse = dict()
purse['money'] = 'six'
purse['candy'] = 3
purse['tissues'] = 75
purse[1.3] = 'andy'
print(purse)
print(purse['candy'])
purse['candy'] = purse['candy'] + 4
print(purse)

x = 'count'
purse[x] = 5
print(purse)
```

```python
# 统一存在个数：if不存在计入，存在+1
counts = dict()
names = ['csev', 'cwen', 'csev', 'zqian', 'cwen']
# in dirct询问的是是否存在这个键
for name in names:
    if name not in counts:
        counts[name] = 1
    else:
        counts[name] = counts[name] + 1
print(counts)
```

```python
# dict.get(key, 0) 将key作为键搜索字典里是否存在，存在返回值，不存在返回给定值
names = ['csev', 'cwen', 'csev', 'zqian', 'cwen']
counts = {}
for name in names:
    counts[name] = counts.get(name, 0) + 1
print(counts) 
```

```python
print('打印字典键和值第一种方法：for循环字典本身后用dict[key]')
count = {'money': 'six', 'candy': 7, 'tissues': 75, 1.3: 'andy', 'count': 5}
for key in count: 
    print(key, count[key])
print('---------')
# list(dict) 字典的键变为列表
print(list(count))
print(count.keys())
print(count.values())
print(count.items()) #返回元素列表的元组

print('打印字典键和值第二种方法：for循环字典的项dict.items()')
for k, v in count.items():
    print(k, v)
```

```python
# Tuples元组 类似列表但不可修改
# 元组不能用sort排序，不能append添加，不能用reverse翻转，顺序保持不变
print('元组第一个用法：同时赋值')
(x, y) = (2, 4)
print(x)
print(y)
```

```python
print('元组的第二个用法：字典的循环变量')
d = {}
d['a'] = 2
d['b'] = 3
print(d)
print(d.items())
for k, v in d.items():
    print(k, v)
```

```python
# 元组比较大小从第一个元素开始逐个比较，一旦符合后面的不再比较
print((0, 1, 2) < (1, 0, -1))
print(('c', 'a', 'd' ) > ('x', 'a', 'd'))
```

```python
# sorted函数排序，返回一个有序列的列表
d = {'a':10, 'b':1, 'c':22}
print(sorted(d))
print(d.items())
sorted(d.items())
```

```python
# 把一个字典的元素根据值从大到小排序
d = {'a':10, 'b':1, 'c':22}
tmp = []
for k, v in d.items():
    tmp.append((v, k))
print(tmp)
tmp = sorted(tmp, reverse = True) # reverse = True 按照从大到小排序
print(tmp)
```

```python
d = {'a':10, 'b':1, 'c':22}
# 列表推导式：[ (v, k) for k, v in d.items() ]
# 将每个元组中的键和值互换位置，得到一个新的列表
print(sorted( [ (v, k) for k, v in d.items() ] ) )
```

# 6. 正则表达式 regular expressions

```python
# re.search('要查询的字符'，查询的位置)检测任意位置匹配
print('使用find查询:')
hand1 = open('mbox.txt')
for line in hand1:
    line = line.rstrip()
    if line.find('From:') >= 0:
        print(line)

print('使用正则表达式re.search查询:')
import re

hand2 = open('mbox.txt')
for line in hand2:
    line = line.rstrip()
    if re.search('From:', line):
        print(line)
```

```python
# 常用的正则表达式方法有：
# re.compile(pattern)：编译一个正则表达式模式，返回一个正则表达式对象1。
# re.match(pattern, string)：从字符串的开头位置匹配一个模式，返回一个匹配对象或None1。
# re.search(pattern, string)：在字符串中的任何位置检测匹配，返回一个匹配对象或None1。
# re.findall(pattern, string)：用包含方式匹配，把所有匹配到的字符放到以列表中的元素返回2。
# re.finditer(pattern, string)：用包含方式匹配，返回一个迭代器，每次迭代返回一个匹配对象2。
# re.sub(pattern, repl, string)：用指定的字符串替换匹配到的子串，返回替换后的字符串3。
# .* 表示任意匹配除换行符（\n、\r）之外的任何单个或多个字符
# (.*?) 表示"非贪婪"模式，只保存第一个匹配到的子串
```

```python
import re
 
print(re.search('www', 'www.runoob.com').span())  # 在起始位置匹配
print(re.search('com', 'www.runoob.com').span())         # 不在起始位置匹配
```

```python
import re
 
phone = "2004-959-559 # 这是一个电话号码"
 
# 删除#后注释
num = re.sub(r'#.*', "", phone)
print ("电话号码 : ", num)
 
# 移除非数字的内容,\D匹配任意非数字
num = re.sub(r'\D', "", phone)
print ("电话号码 : ", num)
```

```python
import re
x = 'From: Using the : character'
y1 = re.findall('^F.+:', x)  #匹配以F开头的一个或多个字符冒号结尾，贪婪匹配
y2 = re.findall('^F.+?:', x) #？非贪婪匹配
print(y1)
print(y2)
```

```python
import re
x = 'From stephen.marquard@uct.ac.za sat Jan 5 09:14:16 2008 da@uct.ac.za'
y1 = re.findall('\S+@\S+', x)
y2 = re.findall('\S@\S', x)
y3 = re.findall('\S+@\S+?', x)
# 行从From开始，要一个空格，要任意数量的字符直到一个@符号，然后提取@后所有的非空白字符，结束提取
y4 = re.findall('^From .*@([^ ]*)', x)
print(y1)
print(y2)
print(y3)
print(y4)
```

```python
# 创建套接字连接
import socket
mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
# 通过internet将该套接字连接到目标域名data.pr4e.org,端口80
mysock.connect(('data.pr4e.org', 80))
# 发送http，encode函数将内部的Unicode转化为UTF-8编码
cmd = 'GET http://data.pr4e.org/romeo.txt HTTP/1.0\n\n'.encode()
mysock.send(cmd)

while True:
    data = mysock.recv(512)
    if (len(data) < 1):
        break
    print(data.decode()) #decode函数将转化为内部格式unicode
mysock.close()
```

导入所需的模块： urllib.request 用于获取网页内容， BeautifulSoup 用于解析HTML。
urllib.parse 和 urllib.error 是Python标准库中的两个模块，用于处理URL和解析URL相关的错误。
urllib.parse 模块提供了一些函数来解析URL、构建URL、编码和解码URL参数等。它包含了一些常用的方法，如 urllib.parse.urlparse() 用于解析URL， urllib.parse.urlencode() 用于编码URL参数， urllib.parse.parse_qs() 用于解析URL参数等.
urllib.error 模块则用于处理与URL相关的错误。它定义了一些异常类，如 urllib.error.URLError 用于表示URL打开或访问出现的错误， urllib.error.HTTPError 用于表示HTTP请求错误等。这两个模块通常与 urllib.request 一起使用，用于处理URL相关的操作和错误处理。

```python
"""导入所需的模块： urllib.request 用于获取网页内容， BeautifulSoup 用于解析HTML。
urllib.parse 和 urllib.error 是Python标准库中的两个模块，用于处理URL和解析URL相关的错误。 
  urllib.parse 模块提供了一些函数来解析URL、构建URL、编码和解码URL参数等。它包含了一些常用的方法，如 urllib.parse.urlparse() 用于解析URL， urllib.parse.urlencode() 用于编码URL参数， urllib.parse.parse_qs() 用于解析URL参数等。 
  urllib.error 模块则用于处理与URL相关的错误。它定义了一些异常类，如 urllib.error.URLError 用于表示URL打开或访问出现的错误， urllib.error.HTTPError 用于表示HTTP请求错误等。 
 这两个模块通常与 urllib.request 一起使用，用于处理URL相关的操作和错误处理。
"""
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
url = input('Enter - ')
# 打开指定的URL，并读取其内容。
html = urllib.request.urlopen(url).read() 
print('读取URL返回文本：', html)
# 使用HTML解析器解析
soup = BeautifulSoup(html, 'html.parser')
print('解析html之后：', soup)
# 找到所有的 <a> 标签，即链接标签。
tags = soup('a')
print('tag是什么', tags)
# 遍历每个链接标签，并打印出其 href 属性的值，即链接地址。如果链接没有 href 属性，则打印 None 。
for tag in tags:
    print(tag.get('href', None))
```

html.parser 、 lxml 和 html5lib 是BeautifulSoup库中可用的解析器类型。它们在解析HTML文档时有一些区别，适用于不同的情况。 
 1.  html.parser ： 
   - 是Python的内置解析器，无需安装额外的库。 
   - 速度较快，适用于大多数HTML文档。 
   - 在处理不规范的HTML时可能会出现解析错误。 
 2.  lxml ： 
   - 是一个高性能的第三方解析器，需要额外安装 lxml 库。 
   - 速度非常快，对于大型HTML文档和复杂的文档结构效果最好。 
   - 支持XPath和CSS选择器，具有强大的功能和扩展性。 
 3.  html5lib ： 
   - 是一个纯Python实现的解析器，需要额外安装 html5lib 库。 
   - 解析HTML文档的容错性很好，可以处理不规范的HTML。 
   - 速度相对较慢，通常比其他解析器慢几倍。 
   - 支持构建DOM树和修改文档结构。 
 选择解析器的最佳情况取决于你的需求和所处理的HTML文档的特点： 
- 如果你处理的是一般的HTML文档，可以选择使用默认的 html.parser ，它是最简单和最快的解析器。 
- 如果你处理的是大型HTML文档或需要较快的解析速度，可以选择 lxml ，它是一个高性能的解析器。 
- 如果你处理的是不规范的HTML文档，或者需要良好的容错性，可以选择 html5lib ，它是一个纯Python实现的解析器，具有良好的容错性。 
 总的来说，如果你不确定选择哪个解析器，可以先尝试使用默认的 html.parser ，然后根据具体需求和性能要求进行调整。

```python
# 导入xml解析模块
import xml.etree.ElementTree as ET
data = """<person>
    <name>Chuck</name>
    <phone type="intl">
        +17343034456
    </phone>
    <email hide="yes"/>
</person>"""
# 解析成树状结构
tree = ET.fromstring(data)
print('name:', tree.find('name').text)
print('Attr:', tree.find('email').get('hide'))
```

```python
import xml.etree.ElementTree as ET

input = '''
<stuff>
  <users>
    <user x="2">
      <id>001</id>
      <name>Chuck</name>
    </user>
    <user x="7">
      <id>009</id>
      <name>Brent</name>
    </user>
  </users>
</stuff>'''

stuff = ET.fromstring(input)
print('解析树结构得到的是')
print(stuff)
lst = stuff.findall('users/user')
print('解析找到标签users/user是一个列表')
print(lst)
print('User count:', len(lst))

for item in lst:
    print('Name', item.find('name').text)
    print('Id', item.find('id').text)
    print('Attribute', item.get('x'))
```

# 9. 使用应用程序接口API

```python
import urllib.request, urllib.parse, urllib.error
import json
import ssl
"""
import ssl  是Python标准库中的一个模块，用于提供安全套接字层（SSL）和传输层安全（TLS）的功能。它提供了用于创建和管理SSL/TLS连接的函数和类。 
 SSL和TLS是用于在网络通信中提供安全性的协议。它们通过对数据进行加密和身份验证来保护数据的机密性和完整性。 ssl 模块提供了与SSL/TLS相关的加密、证书验证和安全连接的功能。 
 在使用 ssl 模块之前，通常需要先通过 ssl.create_default_context() 创建一个SSL上下文对象，然后可以使用该对象进行安全连接的创建和管理。例如，可以使用 ssl.wrap_socket() 函数将普通的套接字包装成安全套接字。 
 需要注意的是，使用 ssl 模块需要正确配置和使用证书，以确保安全连接的有效性和信任
 """

api_key = False
# If you have a Google Places API key, enter it here
# api_key = 'AIzaSy___IDByT70'
# https://developers.google.com/maps/documentation/geocoding/intro

if api_key is False:
    api_key = 42
    serviceurl = 'http://py4e-data.dr-chuck.net/json?'
else :
    serviceurl = 'https://maps.googleapis.com/maps/api/geocode/json?'

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

"""
params是一个包含键值对的字典。urllib.parse.urlencode函数将字典中的键值对转换为URL编码的字符串。键值对之间用 & 连接，键和值之间用 = 连接。空格被转换为 + 符号。
import urllib.parse
params = {'name': 'John Doe', 'age': 25, 'city': 'New York'}
encoded_params = urllib.parse.urlencode(params)
print(encoded_params)
>>>name=John+Doe&age=25&city=New+Yorkv
"""

while True:
    address = input('Enter location: ')
    if len(address) < 1: break

    parms = dict()
    parms['address'] = address
    if api_key is not False: parms['key'] = api_key
    url = serviceurl + urllib.parse.urlencode(parms)

    print('Retrieving', url)
    uh = urllib.request.urlopen(url, context=ctx)
    data = uh.read().decode() # url来自python外部是UTF-8格式，利用decode函数转化为python内unicode格式
    print('Retrieved', len(data), 'characters')

    try:
        js = json.loads(data)
    except:
        js = None

    if not js or 'status' not in js or js['status'] != 'OK':
        print('==== Failure To Retrieve ====')
        print(data)
        continue

    print(json.dumps(js, indent=4)) # 将js对象转换为JSON字符串，并使用4个空格进行缩进
   
    # 经度：Longitude ，简写Lng； 纬度： Latitude， 简写Lat。
    lat = js['results'][0]['geometry']['location']['lat']
    lng = js['results'][0]['geometry']['location']['lng']
    print('输出经纬度：', '纬度lat', lat, '经度lng', lng)
    location = js['results'][0]['formatted_address']
    print('输出完整地址：', location)
    break
```
